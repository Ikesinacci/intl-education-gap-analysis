BEYOND BORDERS: NAVIGATING THE ASPIRATIONS AND REALITIES OF NIGERIAN STUDENTS PURSUING INTERNATIONAL EDUCATION

Problem Statement:
The purpose of this analysis is to explore the motivations, challenges, and realities faced by Nigerian students pursuing education abroad. Using data gathered from respondents during the The International Education (TINE) Africa Summit 2024 held at Sheraton Hotels, Ikeja, we aimed to uncover trends, key insights, and actionable findings that reflect largely, their expectations/aspirations and lived experiences.

IMPORTING LIBRARIES AND DEPENDENCIES (Integrating Survey Spreadsheets)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

# Defining and initializing Google Sheets
def load_google_sheet(sheet_url):
    sheet_id = sheet_url.split('/')[5]
    csv_url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv"
    return pd.read_csv(csv_url)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

LOADING SURVEY DATA

```python
student_data_url = "https://docs.google.com/spreadsheets/d/1DxnVb8Z__bt3--y5NjvNh7T_rl0XhXhW9runsD9yY40/edit?usp=sharing"
stakeholder_data_url = "https://docs.google.com/spreadsheets/d/1YLYC26KPbA_PlByKDS1nwcUjc5MZhQ30_nAOHpBuAU0/edit?usp=sharing"

student_df = load_google_sheet(student_data_url)
stakeholder_df = load_google_sheet(stakeholder_data_url)

print("Student Data:")
display(student_df.head())

print("Stakeholder Data:")
display(stakeholder_df.head())

student_df = student_df.dropna(how='all')  
stakeholder_df = stakeholder_df.dropna(how='all')
```

METHODOLOGY

1. Data Collection:
Responses were collected during the TINE Africa Summit and Awards Conference 2024. The participants included Nigerian students aspiring to pursue international education, education agency representatives, professionals, and counselors within the international education sector. Data was gathered through structured surveys distributed on the first day of the summit.

2. Data Cleaning:
The collected data was examined for missing or redundant entries, which were addressed to ensure accuracy and completeness.

3. Analysis:
On the first day of the summit, printed questionnaires were distributed to attendees, primarily consisting of aspiring students and professionals. On the second day, Google Forms were made available via QR codes to capture additional responses from summit attendees, specifically stakeholders in the international education industry. A total of 99 responses were collected and recorded for analysis. Responses from the Google Forms were also logged for further processing. Data analysis was conducted using Jupyter Notebook, where the responses were coded for deeper insights. Key areas of focus included motivations for studying abroad, challenges faced, countries of interest, and the affordability of international education.

4. Visualization:
Visual aids such as charts and graphs were created to clearly present key insights from the data, enhancing the interpretation and understanding of the findings.

5. Report:
The final report will provide a comprehensive overview of Nigerian students’ aspirations, the challenges they face, and the support they require to succeed in their international education journey.


Key Areas of Analysis exploring the Aspirations of Nigerian Students Pursuing International Education Overseas:
1. Motivations for studying abroad
2. Challenges encountered
3. Preferred countries and institutions
4. Affordability of Institutions

1. Motivations for studying abroad

```python
# Analysis of motivations for studying abroad
motivations_col = '1. What is motivating you to pursue international education?'
motivations_count = student_df[motivations_col].value_counts()

# Mapping options to descriptive labels
motivations_labels = {
    'a': 'Career advancement',
    'b': 'Quality education',
    'c': 'Exposure to a new culture',
    'd': 'Access to advanced educational facilities'
}

motivations_count.index = motivations_count.index.map(motivations_labels)

plt.figure(figsize=(7, 3))
sns.barplot(x=motivations_count.index, y=motivations_count.values, palette="viridis")
plt.title("Motivations for Studying Abroad")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.xlabel("Motivations")
plt.show()
```

2. Challenges encountered

```python
challenges_col = '6. What are your primary concerns about studying abroad?'
challenges_count = student_df[challenges_col].value_counts()

challenges_labels = {
    'a': 'Funding challenges',
    'b': 'Cultural adjustment difficulties',
    'c': 'Academic workload',
    'd': 'Post-graduation employment opportunities'
}

challenges_count.index = challenges_count.index.map(challenges_labels)

plt.figure(figsize=(7, 3))
sns.barplot(x=challenges_count.index, y=challenges_count.values, palette="rocket")
plt.title("Challenges Encountered by Nigerian Students Abroad")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.xlabel("Challenges")
plt.show()
```

3. Preferred countries and institutions

```python
countries_col = '20. What is your preferred study destination?' 
countries_count = student_df[countries_col].value_counts()

plt.figure(figsize=(9, 3))
sns.barplot(x=countries_count.index, y=countries_count.values, palette="coolwarm")
plt.title("Top Countries Preferred for International Education")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.show()
```

4. Affordability of Institutions

```python
funding_col = '9. How affordable do you consider the tuition fees for international education in your preferred study destination?'
funding_count = student_df[funding_col].value_counts()

funding_labels = {
    'a': 'Very affordable',
    'b': 'Somewhat affordable',
    'c': 'Not very affordable',
    'd': 'Not affordable at all'
}

funding_count.index = funding_count.index.map(funding_labels)

plt.figure(figsize=(7, 3))
sns.barplot(x=funding_count.index, y=funding_count.values, palette="Set2")
plt.title("Affordability of International Education")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.xlabel("Affordability")
plt.show()
```

2. Financial Realities of Studying Abroad

```python
financial_realities_col = '5. What are the financial realities of studying abroad for Nigerian students?'

financial_realities_count = stakeholder_df[financial_realities_col].value_counts()

plt.figure(figsize=(6, 2))
sns.barplot(x=financial_realities_count.index, y=financial_realities_count.values, palette="rocket")
plt.title("Financial Realities of Studying Abroad for Nigerian Students")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.xlabel("Financial Realities")
plt.show()
```

3. Social Integration Experiences of Nigerian Students

4. Policies or Regulations Hindering Nigerian Students’ Access to International Education

```python
policies_col = "10. What policies or regulations in Nigeria or destination countries hinder Nigerian students' ability to pursue international education?"

policies_count = stakeholder_df[policies_col].value_counts()

plt.figure(figsize=(6, 2))
sns.barplot(x=policies_count.index, y=policies_count.values, palette="rocket")
plt.title("Barriers to International Education for Nigerian Students")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.xlabel("Policies & Restrictions")
plt.show()
```

SUMMARY OF INSIGHTS: LIVED REALITIES OF NIGERIAN STUDENTS PURSUING INTERNATIONAL EDUCATION

1. Major Challenges Faced by Nigerian Students:
- The primary challenges identified by Nigerian students studying abroad include cultural adaptation, language barriers, and financial constraints, with academic rigor also being a significant concern. These challenges highlight the need for better support systems to ease the transition for students abroad.

2. Financial Realities of Studying Abroad:
- Respondents emphasized that high tuition fees and living expenses are significant barriers for Nigerian students, with many students relying on financial assistance from family or loans. Additionally, fluctuating exchange rates and the lack of adequate financial support make international education a daunting prospect for many.

3. Social Integration of Nigerian Students:
- Nigerian students typically integrate into social networks abroad through student organizations, cultural groups, and online communities. These networks are crucial for overcoming isolation and enhancing the overall student experience, although there remains room for improved support in fostering these connections.

4. Barriers from Policies and Regulations:
- Both Nigerian government policies and host country regulations were identified as major hindrances to Nigerian students’ ability to pursue international education. These include visa restrictions, bureaucratic delays, and limited access to scholarships, which create obstacles for students aspiring to study abroad.
